```python
# There exists four basic variables in python: numeric, string, list, and tuple variables.
```

```python
# This is how you define a string variable:
```

```python
organism = "Mycoplasma genitalium"
print (organism + " " + "is a common contaminant in laboratory settings.")
```

```python
# This is how you define a numeric variable (integer & float):
```

```python
number_of_bps = 1242415
print ("number of base pairs: ", number_of_bps)

percent_A = 24.7
percent_T = 23.6
```

```python
# This is how you define a list variable:
```

```python
percents_AGCT = [percent_A, 26.0, 25.7, percent_T]
print ("[A], [G], [C], [T]=", percents_AGCT)
```

```python
# You may perform computations on these variables by defining the operations as a variable:
```

```python
ratio_AT = percent_A / percent_T
print(ratio_AT)
```

```python
# List vars may be referenced but you may only call on the integers or floats defined in the list
```

```python
# Otherwise, just reference the variable you already defined in the list
```

```python
ratio_GC = percents_AGCT[1]/percents_AGCT[2]
print(ratio_GC)
```

```python
# Note how there [1] and [2] are referenced despite it being the second and fourth in the list, this is because
```

```python
# Variables in the first and fourth position are not considered when using these brackets []
```

```python
# Tuple variables just combine variables into one variable
```

```python
Mycoplasma = (organism, ratio_AT, ratio_GC)
print(Mycoplasma)
```

```python
# Note that if you don't add the asterik the sep function won't work as it only works on multiple arguments which * splits
# Modern python uses f strings for readability in reporting statistics
pi = 3.14159
print (f"Pi is approximately {pi:.2f}")
# .2f indicates 2 decimal places with width and alignment
print (f"Pi is approximately {pi:100.2f}")
print (f"Pi is approximately {pi:<10.2f}")
```
