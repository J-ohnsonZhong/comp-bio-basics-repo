```python
import math

# integers are exact, floats are approximate when used in math, python rounds and then proceeds with math

two_pi = 2.0*math.pi
print (two_pi)
print(math.sin(two_pi))

# note how this result should be approximately 0 but because of rounding error it gives a negative number
```

```python
# strings can be indexed, you may pull the character of each string, you may also go negative with the last character
my_dogs_name = "Spark"
dogs_first_letter = my_dogs_name[0]
dogs_second_letter = my_dogs_name[1]
dogs_third_letter = my_dogs_name[2]
dogs_fourth_letter = my_dogs_name[3]
dogs_fifth_letter = my_dogs_name[-1]
print(dogs_first_letter + dogs_second_letter + dogs_third_letter + dogs_fourth_letter + dogs_fifth_letter + " is my dog.")

# You can define a variable then modify it by redefining it as itself but with a function which offen looks like var.f()

DNA_seq = 'atgagggagt...gatt'
print(DNA_seq)
DNA_seq = DNA_seq.upper()
print(DNA_seq)

# strings can be sliced such that you can define a set of numbers with a lower [0] and upper bound (excluding the upper bound) [3] in [0:3] 
# if you wish to include the upper bound simply leave it blank

first_codon = DNA_seq[0:3]
last_codon = DNA_seq[-3:]
print(first_codon+' '+last_codon)
```

```python
# lists follow similar rules to strings when calling items
stop_codons = ['TAA', 'tAG']
print(stop_codons)

first_stop_codon = stop_codons[0]
print(first_stop_codon)

# While strings are immutable, elements in a list are not immutable

stop_codons[1] = "TAG"
print(stop_codons[1])

# You may add to a list and also determine how many number of items are in a list
stop_codons.append("TGA")
number_of_stop_codons = len(stop_codons)
print(len(stop_codons))

# strings are like a read-only file, lists are editable files, you may change between the two
DNA_seq = ''.join(stop_codons)
print (DNA_seq)
DNA_list = list(DNA_seq)
print (DNA_list)

# As review, we can make this print like a string while still being a list using * as a list unpacker, then sep = '' to remove the spaces

print (*DNA_list, sep='')

# Like a string, you may also slice a list following the same rules

second_codon = DNA_list[3:6]

print(second_codon)

# You may copy a list too so there exists two separate instances of a list where editing one does not change the other
DNA_list_duplicate = DNA_list.copy()
print(DNA_list_duplicate)

# You may add or delete elements in a list
DNA_list_duplicate.insert(5, "?")
print("insert" + ''.join(DNA_list_duplicate))
DNA_list_duplicate.pop(5)
print("popped " + ''.join(DNA_list_duplicate))
```

```python
# Tuples are used to group together data that should not change as opposed to a string which is a singular piece of information

Histidine = ("H", "CAT", "CAC")
Lysine = "K", "AAA", "AAG"
Arginine = ('R', 'CGT', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG')

print (Histidine, Arginine, Lysine)

# Tuples can be sorted into a list, think adjustable organization with fixed files
basic = [Histidine, Lysine]
basic.append(Arginine)
print (basic)

# Nested tuples also exist

basic2 = Histidine, Lysine, Arginine
print(basic2)

# Note that the output print strings show that [] is the data stored as a list, () means the data is stored as a tuple
# You may access the elements in a list of a tuple just like a list
Lys = basic[1]
print (Lys)

# Lets access the list of elements in the tuple
His_codons = basic[0][1:]
print(His_codons)

# You can unpack a tuple by designating variables to each element 
codon1, codon2 = His_codons
print (codon1)
print (codon2)

protein_seq = basic[0][0] + basic[1][0] + basic[2][0]
print(protein_seq)
```

```python
# Dictionaries are lists but instead of reading the sequence with numbers [0], it uses "keys" to reference values
restriction_enzymes = {'EcoRI' : 'GAATC',
                       'AluI' : 'AGCT',
                       'NotI' : 'GCGGCGC',
                       'TaqI' : 'TCGA'
                      }
print(restriction_enzymes) # Printing the dictionary gives both key and value

# In order to get only keys or values you would convert the dictionary to a list for .values() or .keys()
keys = restriction_enzymes.keys() #The .keys() function returns a view object which is an inline object kind of, meaning its neither a tuple nor list
print(keys) #notice how it prints as a view object and not as a list or tuple
keys = list(restriction_enzymes.keys()) #now this is a list
print (keys)

# Lets say you want to see if a key is in a dictionary
mykey = "crispr"
check = mykey in restriction_enzymes

# Now lets see if we can get a value from a dictionary with its key
EcoRI_RS = restriction_enzymes['EcoRI'] # Note, use this if you always expect a key to exist or missing info is unacceptable, use .get() if its acceptable
print(EcoRI_RS)

# Dictionaries are mutable, here is how you add to them

restriction_enzymes['EcoRV'] = 'GATATC' # likewise, if you expect this key to not exist then use this function, otherwise use update to handle redundancy

print(restriction_enzymes)
# .update() is also better for adding multiple keys and values all at the same time

# Now this is how you delete from a dictionary

del restriction_enzymes['EcoRV']
print(restriction_enzymes)
```
